```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn import datasets
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn import tree
from sklearn.cross_validation import KFold, cross_val_score
from sklearn.cross_validation import train_test_split
from sklearn.metrics import roc_curve # ROC Curves
from sklearn.metrics import auc
%matplotlib inline
pd.set_option('display.max_columns',300)

df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data', header=None, sep=', ')


df = df.rename(columns={0:'age',1:'workclass',2:'fnlwgt',3:'education',4:'education-num',5:'marital-status',6:'occupation',7:'relationship',8:
                'race',9:'sex',10:'capital-gain',11:'capital-loss',12:'hours-per-week',13:'native-country',14:'income'})
```

```python
df2 = pd.get_dummies(df)
del df2['income_>50K']
df2
```

```python
features = df2.columns[:-1]
print(features)

X = df2[features]
y = df2["income_<=50K"]

X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=13, test_size=0.2)

logreg = LogisticRegression()
logreg.fit(X_train,y_train)
```

```python
y_pred = logreg.predict(X_test)

print('LogReg %s' % metrics.accuracy_score(y_test, y_pred))
```

```python
fpr, tpr, _ = roc_curve(y_pred, y_test)
auc_nn = auc(fpr, tpr)
```

```python
fig, ax = plt.subplots(figsize=(10,10))

plt.plot(fpr, tpr, label = "Income Prediction ROC curve" % auc_nn)

plt.plot([0,1],[0,1],'k--', color='black')
plt.plot([0,0],[1,0],'k--', color='black')
plt.plot([1,0],[1,1],'k--', color='black')

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve Income Prediction')
plt.legend(loc='lower right')
plt.show()
```
